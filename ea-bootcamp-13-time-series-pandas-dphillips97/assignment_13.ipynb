### Assignment 13: Time Series with Pandas
#### Analysis of Precipitation and Stream Discharge in Boulder, CO, During the 2013 Flood

This notebook displays graphical precipitation values for Boulder and flow rate for the USGS flow station at Boulder Creek at N 75th St. A high-level summary of the workflow is:
1. Download data using earthpy library
2. Load datasets into pandas
3. "Extract" datasets for relevant time periods, if necessary
4. Resample datasets by desired time period
5. Graph results
6. Interpret findings

```python
# Import packages
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import seaborn as sns
import pandas as pd
import earthpy as et

# Handle datetime conversions b/w matplotlib & pandas 
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Download folder containing all datasets
et.data.get_data('colorado-flood')
```

```python
# Set working directory (home/earth-analytics)
work_dir = os.path.join(et.io.HOME, "earth-analytics")

# Go to working dir if it exists, print message
if os.path.exists(work_dir):
    os.chdir(work_dir)
    print(f"Path changed to {work_dir}")
else:
    print("Path does not exist")
```

```python
# Set path to precipitation data
precip_path = os.path.join(work_dir, "data", 
                           "colorado-flood",
                           "precipitation", 
                           "805333-precip-daily-1948-2013.csv")

# Set path to water discharge/flow data
flow_path = os.path.join(work_dir, "data", 
                         "colorado-flood",
                         "discharge", 
                         "06730200-discharge-daily-1986-2013.csv")
```

```python
# Load precipitation dataset to pandas df
# Index is set to datetime object 
# na values (999.99) are replaced with NaN
precip = pd.read_csv(precip_path,
                    parse_dates=['DATE'],
                    index_col=['DATE'],
                    na_values=['999.99'])
```

```python
# Get precip values from Aug 01 2013 to Oct 01 2013 
precip_flood = precip['2013-08-01':'2013-10-01']

# Resample to find daily sum
precip_flood_day = precip_flood.resample('D').sum()
```

```python
# Get all precip values for 2013
precip_2013 = precip["2013"]

# Resample to find monthly max:
# find daily max then monthly max
precip_2013_mon_max = precip_2013.resample('D').max().resample('M').max()
```

```python
# Get precip values for all years by month
precip_all_years = precip.resample('M').sum()
```

```python
"""Plot precipitation values"""
# Create figure and 3 plot spaces
fig, (ax1, ax2, ax3) = plt.subplots(3,1,figsize=(10, 15))

# Set title for entire figure
fig.suptitle("Fig 1. Precipitation Analysis for Boulder, CO", 
             fontsize=18)

"""For daily values in time period """
# Add x-axis and y-axis
ax1.plot(precip_flood_day.index.values,
           precip_flood_day['HPCP'],
           color='blue')

# Format x axis, labels like "Jan-01"
date_form = DateFormatter("%b-%d")

# Set format for x axis
ax1.xaxis.set_major_formatter(date_form)

# Set title and axes labels 
ax1.set(xlabel="Date",
        ylabel="Precipitation (inches)",
        title="Daily Total Precipitation, Aug to Oct 2013")


"""For Monthly Max in 2013"""
ax2.plot(precip_2013_mon_max.index.values,
           precip_2013_mon_max['HPCP'],
           color='green')

# Format x axis, labels like "Jan-01"
date_form_2 = DateFormatter("%b-%d")

# Set format for x axis
ax2.xaxis.set_major_formatter(date_form_2)

# Show all months
ax2.xaxis.set_major_locator(mdates.MonthLocator())

# Set title and labels for axes
ax2.set(xlabel="Month",
        ylabel="Precipitation (inches)",
        title="Monthly Precipitation in 2013")


"""Monthly Totals for all years"""
ax3.plot(precip_all_years.index.values,
         precip_all_years['HPCP'],
         color='orange',
         linewidth=0.5)

# Set title and labels for axes
ax3.set(xlabel="Year",
        ylabel="Precipitation (inches)",
        title="Yearly Total Precipitation")

# Show figure with 3 subplots
plt.show()
```

```python
# Load water discharge dataset into pandas df
# Set datetime object to index
# No null values in this dataset
flow = pd.read_csv(flow_path,
                   parse_dates=['datetime'],
                   index_col=['datetime'])
```

```python
# Get flow data for Aug 01 2013 to Oct 01 2013
flow_flood = flow['2013-08-01':'2013-10-01']

# Resample to find sum per day
flow_flood_day = flow_flood.resample('D').sum()
```

```python
# Get flow data for 2013
flow_2013 = flow['2013']

# Resample to find daily max
# Find daily max then find monthly max
flow_2013_mon_max = flow_2013.resample('D').max().resample('M').max()
```

```python
# Find monthly total for all years by month
flow_all_years = flow.resample('M').sum()
```

```python
"""Plot flow rate values"""
# Create figure and 3 plot spaces
fig, (ax1, ax2, ax3) = plt.subplots(3,1,figsize=(10, 15))

# Set title for entire figure
fig.suptitle("Fig 2. Discharge Analysis for Boulder Creek", 
             fontsize=18)

"""For daily values in time period """
# Add x-axis and y-axis
ax1.plot(flow_flood_day.index.values,
         flow_flood_day['disValue'],
         color='blue')

# Format x axis, labels like "Jan-01"
date_form = DateFormatter("%b-%d")

# Set format for x axis
ax1.xaxis.set_major_formatter(date_form)

# Set title and labels for axes
ax1.set(xlabel="Date",
        ylabel="Flow rate (cubic ft per second)",
        title="Daily Total Discharge, Aug to Oct 2013")


"""For Monthly Max in 2013"""
ax2.plot(flow_2013_mon_max.index.values,
         flow_2013_mon_max['disValue'],
         color='green')

# Format x axis, labels like "Jan-01" 
date_form_2 = DateFormatter("%b-%d")

# Set format for x axis
ax2.xaxis.set_major_formatter(date_form_2)

# Show tick for each month
ax2.xaxis.set_major_locator(mdates.MonthLocator())

# Set title and labels for axes
ax2.set(xlabel="Month",
        ylabel="Flow rate (cubic ft per second)",
        title="Monthly Max Flow Rate in 2013")


"""Monthly Totals for all years"""
ax3.plot(flow_all_years.index.values,
         flow_all_years['disValue'],
         color='orange',
         linewidth=0.5)

# Set title and labels for axes
ax3.set(xlabel="Year",
        ylabel="Flow rate (cubic ft per second)",
        title="Yearly Total Flow Rate")

# Show plot with 3 subplots
plt.show()
```

## Summary

#### Precipitation
- The highest total precipitation values in the time period were 7.3 inches on Sep 12, 6.4 inches on Sep 11, and 1.6 inches on Sep 15. There was a significant spike in values on Sep 11 and 12; there was a smaller jump in values on Sep 15. All other precipitation values were under 1 inch in the time period.
- The monthly maximum precipitation for October, 2.2 inches, was by far the largest value for all months in 2013. The second largest maximum was in July at 0.3 inches.
- There is significant variation in yearly total precipitation from 1948 to 2013. The total monthly precipitation was highest since 1948 in Sep 2013, when the yearly total precipitation was 17.7 inches.

#### Discharge
- Daily total flow rate spiked to 4770 cubic feet per second (cfs) on Sep 12 and 4640 cfs on Sep 13. After the elevated rates during the flood, the flow rate gradually decresed though it still remained at elevated levels by October 1 2013. The flow rate on Sep 11 was only 349 cfs.
- The monthly maximum flow rate for Sep, nearly 5000 cubic feet per second, was the largest value for all months in 2013. The second largest value was in June and was 629 cfs.
- There is significant variation in yearly total precipitation from 1986 to 2013. Interestingly, the largest flow values did not occur during the 2013 flood but in June 1995 (26048 cfs versus 25844 cfs, respectively).

#### Lessons Learned
- During the 2013 flood, the precipitation values exceeded any other year on record. The flow rate in 2013 was the second highest since 1986. Precipitation and discharge were both extremely high in Sep 2013, the month of Boulder's major flooding.
